#Importing Libraries

```
import tensorflow as tf
import os
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing import image
from keras.models import load_model
import tensorflow.python.keras
import numpy as np
```

#Load and Test the Model

```
model = tensorflow.python.keras.models.load_model('/content/drive/My Drive/Colab Notebooks/SmartInternz/ChestXray/my_model.h5')
```

```
#ImageDataGenerator is used for rescaling the testing images
test_datagen = ImageDataGenerator(rescale=1./255)
```

```
test_generator = test_datagen.flow_from_directory(
        '/content/drive/My Drive/Colab Notebooks/SmartInternz/ChestXray/chest_xray/test',   #Paste the path of testing folder here
        target_size=(600, 600),
        color_mode="rgb",
        class_mode='binary',
        batch_size= 4)
```

```
#Prediction for the test set.
predict = model.predict_generator(test_generator,steps = np.ceil(879/4))
```

```
predict
```

```
#Classwise prediction
model.predict_classes(test_generator)
```

```
#Checking the the loss and accuracy for the testing set
acc = model.evaluate(test_generator)
```
